```python
%matplotlib notebook
import pandas as pd
import numpy as np
```

```python
파일위치경로 = r'C:/Users/user/Desktop/구청별 업무추진비(엑셀)/'
구청목록 = [ 
    '종로구','중구','용산구','성동구','광진구','동대문구','중랑구','성북구','강북구','도봉구','노원구',
    '은평구','서대문구','마포구','양천구','강서구','구로구','금천구','영등포구','동작구','관악구','서초구',
    '강남구','송파구','강동구','서울특별시'
            ]   #26개
극단값제거 = 500000
결재방법_변환키 ={'-':'카드', '현금':'현금', '계좌이체':'현금', '무통장입금':'현금', '법인카드':'카드', '카드':'카드', '국내':'카드'}
```

```python
구청별금액사전=dict()

for count, 구청이름 in enumerate(구청목록):
    try:
        구청금액 = pd.read_excel(파일위치경로+구청이름+'.xlsx',index_col='연번',parse_cols='A,J')
        try:
            구청금액 = 구청금액.astype(str)['집행금액'].apply(lambda x : x.replace(',','').replace(' ','')).astype(np.int64)
        except AttributeError:
            pass
        #구청금액 = 구청금액[구청금액<극단값제거].reset_index(drop=True)
        구청별금액사전[구청이름] = 구청금액.reset_index(drop=True)
    except FileNotFoundError:
        pass
    except:
        print(구청이름)
```

```python
구청별금액프레임 =pd.DataFrame()
구청식별코드 = dict()
count =0
for 구청이름 in 구청목록:
    try:
        구청별금액프레임 = pd.concat([구청별금액프레임,구청별금액사전[구청이름]],ignore_index=True,axis=1)
        구청식별코드[count] = 구청이름
        count+=1
    except KeyError:
        pass
```

## 구분별 기초통계

```python
구청별금액프레임 = pd.DataFrame()
for 구청이름 in 구청목록:
    try:
        구청별금액프레임=구청별금액프레임.append(pd.read_excel(파일위치경로+구청이름+'.xlsx',index_col='연번',parse_cols="A,B,D,I,J"))
    except FileNotFoundError:
        pass
    except:
        print(구청이름)
구청별금액프레임.집행금액 = 구청별금액프레임.집행금액.astype(str).apply(lambda x : x.replace(',','').replace(' ','')).astype(np.float64)
```

```python
구청별금액프레임.결재방법=구청별금액프레임.결재방법.apply(lambda x : 결재방법_변환키[x])
```

```python
자치구별_통계 = 구청별금액프레임.groupby('자치구').agg([len,sum,np.mean])
```

```python
구분별_결재별_통계 = pd.pivot_table(구청별금액프레임,columns=['자치구','구분','결재방법'],aggfunc=[len,np.sum,np.mean])
```

```python
결재별_통계 =pd.pivot_table(구청별금액프레임[['자치구','결재방법','집행금액']],columns=['자치구','결재방법'],aggfunc=[len,np.sum])
```
