```python
구청별=dict()

for 구청이름 in 구청목록:
    try:
        임시 = pd.read_excel(파일위치경로+구청이름+'(정제).xlsx')
        임시 = 임시[임시['집행일시']>=일정시점_이후]
        구청별[구청이름] = 임시
    except FileNotFoundError:
        print('파일이 없습니다:'+구청이름)
    except :
        print(구청이름)
```

## 꺽기

```python
for 임시_구청이름, 각_프레임 in 구청별.items():
    각_프레임.시분값 = 각_프레임.시분값.astype(str)
    구분 =list(set(각_프레임.구분))
    직접확인_프레임 = pd.DataFrame(columns=칼럼)
    간접확인_프레임 = pd.DataFrame(columns=칼럼)
    for 임시_구분_이름 in 구분:
        임시_구분 =각_프레임[각_프레임.구분 == 임시_구분_이름]
        일시별_금액합계 =임시_구분.groupby('집행일시').sum()
        임시_프레임 =pd.DataFrame(columns=칼럼)
        for 임시_일시 in list(일시별_금액합계[일시별_금액합계.집행금액 >300000].index):
            #print(각_프레임[각_프레임.집행일시 == i])
            임시_프레임 =각_프레임[각_프레임.집행일시 == 임시_일시]
            임시_프레임 = 임시_프레임[임시_프레임.구분 ==임시_구분_이름]
            직접확인_프레임 =직접확인_프레임.append(임시_프레임)  #뽑힌거 일단 모두 저장

        임시_장소종류 = set(임시_프레임.집행장소)
        for 임시_장소 in 임시_장소종류:
            if 임시_프레임[임시_프레임.집행장소 == 임시_장소].집행일시.count() > 1:
                간접확인_프레임 = 간접확인_프레임.append(임시_프레임[임시_프레임.집행장소 == 임시_장소])
    if not 직접확인_프레임.empty:
        직접확인_프레임.to_excel(r"C:/Users/user/Desktop/구청별 업무추진비(엑셀)/꺽기/하루에_일정비용이상사용_직접확인필요/"+임시_구청이름+"_30만원이상.xlsx")
    if not 간접확인_프레임.empty:
        간접확인_프레임.to_excel(r"C:/Users/user/Desktop/구청별 업무추진비(엑셀)/꺽기/하루에_일정비용이상사용_상호명같은것만분류/"+임시_구청이름+"_30만원이상.xlsx")
```

## 휴일이용

```python
for 임시_구청이름, 각_프레임 in 구청별.items():
    각_프레임[(각_프레임.요일 =='일') | (각_프레임.요일 =='토')].to_excel(파일위치경로+'/휴일사용//'+임시_구청이름+'(토_일).xlsx')
```

```python
휴일이용_프레임 =pd.DataFrame(index =['금', '목', '수', '월', '일', '토', '화', '휴일이용비율', '합계'])
for 임시_구청이름, 각_프레임 in 구청별.items():
    임시_프레임 =각_프레임.groupby('요일')['자치구'].count()
    임시_합계 =임시_프레임.sum()
    임시_일 = 0
    임시_토 = 0
    try:
        임시_토 = 임시_프레임['토']
    except KeyError:
        pass
    try:
        임시_일 = 임시_프레임['일']
    except KeyError:
        pass
    임시_프레임['휴일이용비율'] = (임시_토+임시_일) / 임시_합계
    임시_프레임['합계']=임시_합계
    임시_프레임.name = 임시_구청이름
    휴일이용_프레임[임시_구청이름] = 임시_프레임
휴일이용_프레임.to_excel(파일위치경로+'/휴일사용//구청별_휴일이용비율.xlsx')
    
        
        
    
```

```python
평일_주말_합계액 = pd.DataFrame(index=['평일','주말'])
for 임시_구청이름, 각_프레임  in 구청별.items():
    임시 = 각_프레임.groupby('요일')
    try:
        임시_주말_토 = 임시.sum()['집행금액'].loc['토']
    except KeyError:
        임시_주말_토 = 0
    try:
        임시_주말_일 = 임시.sum()['집행금액'].loc['일']
    except KeyError:
        임시_주말_일 = 0
    임시_주말 = 임시_주말_토+임시_주말_일
    임시_평일 = 임시.sum()['집행금액'].loc[['월','화','수','목','금']].sum()
    평일_주말_합계액[임시_구청이름] = [임시_평일,임시_주말]
    평일_주말_합계액
```

```python
평일_주말_합계액.loc['합계'] = 평일_주말_합계액.sum()
평일_주말_합계액.T.to_excel(파일위치경로+'/휴일사용/평일_주말_합계액.xlsx')
```
