```python
구청별=dict()

for 구청이름 in 구청목록:
    try:
        구청별[구청이름]=pd.read_excel(파일위치경로+구청이름+'(정제).xlsx',index_col='연번')
    except FileNotFoundError:
        print('파일이 없습니다:'+구청이름)
    except :
        print(구청이름)
```

## 꺽기

```python
for 임시_구청이름, 각_프레임 in 구청별.items():
    각_프레임.시분값 = 각_프레임.시분값.astype(str)
    구분 =list(set(각_프레임.구분))
    직접확인_프레임 = pd.DataFrame(columns=칼럼)
    간접확인_프레임 = pd.DataFrame(columns=칼럼)
    for 임시_구분_이름 in 구분:
        임시_구분 =각_프레임[각_프레임.구분 == 임시_구분_이름]
        일시별_금액합계 =임시_구분.groupby('집행일시').sum()
        임시_프레임 =pd.DataFrame(columns=칼럼)
        for 임시_일시 in list(일시별_금액합계[일시별_금액합계.집행금액 >300000].index):
            #print(각_프레임[각_프레임.집행일시 == i])
            임시_프레임 =각_프레임[각_프레임.집행일시 == 임시_일시]
            임시_프레임 = 임시_프레임[임시_프레임.구분 ==임시_구분_이름]
            직접확인_프레임 =직접확인_프레임.append(임시_프레임)  #뽑힌거 일단 모두 저장

        임시_장소종류 = set(임시_프레임.집행장소)
        for 임시_장소 in 임시_장소종류:
            if 임시_프레임[임시_프레임.집행장소 == 임시_장소].집행일시.count() > 1:
                간접확인_프레임 = 간접확인_프레임.append(임시_프레임[임시_프레임.집행장소 == 임시_장소])
    if not 직접확인_프레임.empty:
        직접확인_프레임.to_excel(r"C:/Users/user/Desktop/구청별 업무추진비(엑셀)/꺽기/직접/"+임시_구청이름+"_직접.xlsx")
    if not 간접확인_프레임.empty:
        간접확인_프레임.to_excel(r"C:/Users/user/Desktop/구청별 업무추진비(엑셀)/꺽기/간접/"+임시_구청이름+"_간접.xlsx")
```
